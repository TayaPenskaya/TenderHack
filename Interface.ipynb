Description: 
- запрос (список слов) 
- пробегаемся по каждому слову из списка: слово 
- исправление без ошибок(слово) -> слово
    - смитана -> сметана
- нормализация(слово) -> слово
    - коту -> кот

```python
import requests
```

```python
from pyaspeller import Word
```

```python
import pymorphy2
```

```python
import re
```

```python
request = str(input())
```

```python
words = re.sub(r'\s+', ' ', request).split(' ')
print(words)
```

```python
def error_correction(word):
    """Corrects errors in a word.
    """
    check_wrong = Word(word)
    if check_wrong.correct:
        corrected_word = word
    else:
        corrected_word = check_wrong.variants[0];
    return str(corrected_word)
```

```python
def normalization(word):
    """Normalize word.
    """
    morph = pymorphy2.MorphAnalyzer()
    parsed_word = morph.parse(word)[0]
    return parsed_word.normal_form
```

```python
def word_processing(word):
    """Correct, normalize and get synonyms of word.
    
    Returns: 
        Dictionary of a correct & normalized word and its synonyms.
    """
    print('Начальное слово : '+ word)
    corrected = error_correction(word)
    print('Слово без ошибок : ' + corrected)
    normalized = normalization(corrected)
    print('Полученный инфинитив от слова: ' + normalized)
```

```python
for word in words:
    word_processing(word)
```

## Part 3.

Description:
- парсим "надежность" по ИНН со сбис
- добавляем колонку с надежностью

```python
import urllib
```

```python
import bs4
```

```python
from bs4 import BeautifulSoup
```

```python
def getReliabilityByInn(inn):
    contragent = "https://sbis.ru/contragents/"
    
    f = urllib.request.urlopen(contragent+str(inn))
    
    contents = f.read()
    
    soup = BeautifulSoup(contents, 'lxml') 
    
    reliability = soup.find("div", {"class": "cCard__Reliability"})
    
    plus = reliability.find("div", {"class": "analytics-ReliabilitySbisRu__smallTextGreen analytics-ReliabilitySbisRu__left"})
    
    minus = reliability.find("div", {"class": "analytics-ReliabilitySbisRu__smallTextRed analytics-ReliabilitySbisRu__left"})

    plus_num = str(plus)[str(plus).rfind('+')+2:str(plus).rfind('<')]
    
    minus_num = str(minus)[str(minus).rfind('-')+2:str(minus).rfind('<')]
        
    f.close()  
    
    print('+' + plus_num)
    
    print('-' + minus_num)
    
    return [plus_num, minus_num]
```

```python
getReliabilityByInn(5032999602)
```

Description: 
- находим по слову список синонимов
- ищем совпадения с синонимами в тексте

```python
import sys 
```

```python
from portYaDict import YandexDictionary
```

```python
def search_syn(word):
    
    # Получаем ответ API на запрос по слову
    dictionary = YandexDictionary(config.API_DICT_KEY)
    output = dictionary.lookup(word, 'ru', 'ru')
    
    # Фильтруем (очищаем) данные
    russian_letters = set("ёйцукенгшщзхъфывапролджэячсмитьбю ")
    russian_text = ""
    for letter in output:
        if letter in russian_letters:
            russian_text += letter    
        else:
            russian_text += ","
    
    # Массив синонимов, возвращаем первые 5
    synonymus = []
    synonymus = list(filter(lambda x: len(x.split(" ")) == 1 and len(x) > 0, list(russian_text.split(','))))
    
    return synonymus[0:5]
```

```python
search_syn("город")
```
